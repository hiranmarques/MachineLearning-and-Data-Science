```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Carregar os dados
data = pd.read_csv('drinks.csv')

# Selecionar as features relevantes para o modelo não supervisionado
X = data[['beer_servings', 'spirit_servings', 'wine_servings']]

# Normalizar os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Escolher o número de clusters (grupos)
num_clusters = 3

# Criar o modelo K-means
kmeans = KMeans(n_clusters=num_clusters, random_state=42)

# Treinar o modelo
kmeans.fit(X_scaled)

# Obter os rótulos dos clusters
labels = kmeans.labels_

# Adicionar os rótulos dos clusters ao DataFrame original
data['cluster'] = labels

# Visualizar os clusters
plt.figure(figsize=(10, 6))
for cluster in range(num_clusters):
    cluster_data = data[data['cluster'] == cluster]
    print(cluster_data.columns)  # Verificar as colunas disponíveis no DataFrame cluster_data
    plt.scatter(cluster_data['beer_servings'], cluster_data['spirit_servings'], label=f'Cluster {cluster}')
plt.xlabel('Beer Servings')
plt.ylabel('Spirit Servings')
plt.title('Drinks')
plt.legend()
plt.show()
```
